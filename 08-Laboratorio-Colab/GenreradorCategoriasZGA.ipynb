```python
!pip install supabase requests pandas

import requests
import time
import datetime
from google.colab import userdata
from supabase import create_client, Client

# ==========================================
# 1. CONFIGURACI√ìN Y M√ìDULOS
# ==========================================

# Lista de m√≥dulos a respaldar. Puedes comentar (#) los que no quieras.
MODULES_TO_BACKUP = {
    # "endpoint_url": "id_field_in_json"
    "contacts": "contact_id",
    "items": "item_id",
    "invoices": "invoice_id",
    "estimates": "estimate_id",
    "salesorders": "salesorder_id",
    "creditnotes": "creditnote_id",
    "purchaseorders": "purchaseorder_id",
    "bills": "bill_id",
    "vendorcredits": "vendor_credit_id",
    "chartofaccounts": "account_id", # Plan de cuentas
    "bankaccounts": "account_id",
    "journals": "journal_id",        # Asientos de diario
    "users": "user_id"
}

# Carga de Secretos
try:
    ZOHO_CLIENT_ID = userdata.get('ZOHO_CLIENT_ID')
    ZOHO_CLIENT_SECRET = userdata.get('ZOHO_CLIENT_SECRET')
    ZOHO_REFRESH_TOKEN = userdata.get('ZOHO_REFRESH_TOKEN')
    SUPABASE_URL = userdata.get('SUPABASE_URL')
    SUPABASE_KEY = userdata.get('SUPABASE_KEY')
except Exception as e:
    print("‚ùå Error cargando secretos.")
    raise e

supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)

# ==========================================
# 2. FUNCIONES HELPER
# ==========================================

def get_access_token():
    url = "https://accounts.zoho.com/oauth/v2/token"
    payload = {
        "refresh_token": ZOHO_REFRESH_TOKEN,
        "client_id": ZOHO_CLIENT_ID,
        "client_secret": ZOHO_CLIENT_SECRET,
        "grant_type": "refresh_token"
    }
    try:
        resp = requests.post(url, data=payload)
        if resp.status_code == 200:
            return resp.json().get("access_token")
        print(f"‚ùå Error Auth: {resp.text}")
        return None
    except Exception as e:
        print(f"‚ùå Error Red: {e}")
        return None

def get_all_organizations(token):
    """Obtiene la lista de todas las organizaciones de la cuenta"""
    headers = {'Authorization': f'Zoho-oauthtoken {token}'}
    r = requests.get("https://www.zohoapis.com/books/v3/organizations", headers=headers)
    if r.status_code == 200:
        return r.json().get("organizations", [])
    print(f"‚ö†Ô∏è Error obteniendo organizaciones: {r.text}")
    return []

def backup_module(token, org_id, module_name, id_key):
    """Respalda un m√≥dulo espec√≠fico completo"""
    print(f"   ‚Ü≥ üìÇ Iniciando m√≥dulo: {module_name.upper()}...")

    headers = {'Authorization': f'Zoho-oauthtoken {token}'}
    page = 1
    has_more = True
    module_count = 0

    while has_more:
        # Paginaci√≥n
        url = f"https://www.zohoapis.com/books/v3/{module_name}?organization_id={org_id}&page={page}&per_page=200"

        try:
            r = requests.get(url, headers=headers)

            # Control de Rate Limit (Si Zoho dice "para un poco")
            if r.status_code == 429:
                print("      ‚è≥ Rate Limit alcanzado. Esperando 10 segundos...")
                time.sleep(10)
                continue # Reintentar la misma p√°gina

            if r.status_code != 200:
                print(f"      ‚ö†Ô∏è Error en p√°g {page}: {r.status_code} - Ignorando resto del m√≥dulo.")
                break

            data = r.json()
            # La lista suele venir en la clave con el nombre del m√≥dulo (ej: "invoices": [...])
            # A veces es singular o plural, intentamos adivinar o usar la key
            items_list = data.get(module_name)
            if items_list is None:
                # Intento de fallback si el nombre de la key es distinto al endpoint
                keys = [k for k in data.keys() if isinstance(data[k], list) and k != 'page_context']
                if keys: items_list = data[keys[0]]

            if not items_list:
                has_more = False
                break

            # PREPARAR LOTE PARA SUPABASE
            records = []
            for item in items_list:
                # Intentar obtener un nombre legible para la columna record_name
                readable_name = item.get('name') or item.get('customer_name') or item.get('invoice_number') or item.get('reference_number') or item.get('date') or "N/A"

                records.append({
                    'org_id': org_id,
                    'module_type': module_name,
                    'zoho_record_id': str(item.get(id_key)), # Usamos el ID map
                    'record_name': str(readable_name),
                    'raw_data': item,
                    'last_synced_at': datetime.datetime.now(datetime.timezone.utc).isoformat()
                })

            # UPSERT A SUPABASE
            # Usamos on_conflict en las 3 columnas clave
            supabase.table('zoho_universal_backup').upsert(records, on_conflict='org_id,module_type,zoho_record_id').execute()

            module_count += len(records)
            print(f"      ‚úÖ P√°g {page}: {len(records)} registros guardados.")

            page += 1
            time.sleep(0.2) # Pausa cort√©s

        except Exception as e:
            print(f"      ‚ùå Error cr√≠tico en backup_module: {e}")
            break

    print(f"   ‚ú® M√≥dulo {module_name} finalizado. Total: {module_count}")

# ==========================================
# 3. ORQUESTADOR PRINCIPAL
# ==========================================

def run_full_backup():
    print("üîÑ --- INICIANDO BACKUP UNIVERSAL ZOHO BOOKS ---")

    token = get_access_token()
    if not token: return

    # 1. Obtener Organizaciones
    orgs = get_all_organizations(token)
    print(f"üè¢ Se encontraron {len(orgs)} Organizaciones.")

    for org in orgs:
        org_id = org['organization_id']
        org_name = org['name']
        print(f"\nüåç PROCESANDO ORGANIZACI√ìN: {org_name} (ID: {org_id})")
        print("="*60)

        # 2. Iterar M√≥dulos
        for module, id_key in MODULES_TO_BACKUP.items():
            backup_module(token, org_id, module, id_key)

    print("\nüèÅ --- PROCESO COMPLETADO EXITOSAMENTE ---")

if __name__ == "__main__":
    run_full_backup()
```
